```python
# CONVERTING MILLISECONDS TO SECONDS IN SPOTIFY
```

```python
track['duration'] = track['duration_ms'].apply(lambda x: round(x/1000))
track.drop('duration_ms',inplace=True,axis=1)
```

```python
track.duration.head()
```

```python
# HEATMAP CORRELATION
```

```python
cm = track.drop(['key','mode','explicit'], axis=1).corr(method = 'pearson')
plt.figure(figsize=(14,6))
map = sns.heatmap(cm, annot = True, fmt = '.1g', vmin=-1, vmax=1, center=0, cmap='inferno', linewidths=1, linecolor='Black')
map.set_title('Correlation Heatmap between Variable')
map.set_xticklabels(map.get_xticklabels(), rotation=90)
```

```python
# take 0.4% of data for regression
```

```python
sam = track.sample(int(0.004 * len(track)))
```

```python
# Regression between loudness and Energy
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sam, y='loudness', x='energy', color='c').set(title='Loudness vs Energy Correlation')
```

```python
# Regression between Popularity and Acousticness
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sam, y='popularity', x='acousticness', color='b').set(title='Popularity vs Acousticness Correlation')
```

```python
# Creating New Column For Year,RealeseDate
```

```python
# track['dates']=track.index.get_level_values('release_date')
track.dates=pd.to_datetime(track.release_date)
years=track.dates.dt.year
```

```python
sns.displot(years, discrete=True, aspect=2, height=5, kind='hist').set(title='Number of songs per year')
```

```python
#change in duration over the years 
```

```python
total_dr = track.duration
fig_dims = (18,7)
fig, ax = plt.subplots(figsize=fig_dims)
fig = sns.barplot(x = years, y = total_dr, ax = ax, errwidth = False).set(title='Years vs Duration')
plt.xticks(rotation=90)
```

```python
#Duration of songs in differnet genres
```

```python
plt.title('Duration of songs in different Genres')
sns.color_palette('rocket', as_cmap=True)
sns.barplot(y='genre', x='duration_ms', data=genre)
plt.xlabel('Duration in milliseconds')
plt.ylabel('Genres')
```
